```python
import numpy as np
import torch
from locationing_dataset import MyLocationingDataSet
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
root_dir = "C:/Users/KZTYLF/Documents/playground/GNN UIs/GNN InstanceSegmentation/Recreating Dataset/locationing_dataset/train/"
pc_dir = root_dir+"point_clouds/"
label_dir = root_dir+"labels/"
dataset_selection_idx = 0
```

```python
myDataset = MyLocationingDataSet(pc_dir, label_dir)
# vertex_coord_list, keypoint_indices_list, edges_list, final_point, final_edges = myDataset[dataset_selection_idx]
graph, label = myDataset[dataset_selection_idx]
```

```python
## Vertex Coord List
print("\nVertex Coord List")
for i,el in enumerate(graph["vertex_coord_list"]):
    print(i, el.shape)

## KeyPoint Indices List
print("\nKeyPoint Indices List")
for i,el in enumerate(graph["keypoint_indices_list"]):
    print(i, el.shape)

## Edge List
print("\nEdge List")
for i,el in enumerate(graph["edges_list"]):
    print(i, el.shape)
```

```python
## Visualizing The Different Layers

colours = ["blue", "green", "yellow", "red" ]
## Individually First
for i,el in enumerate(graph["vertex_coord_list"]):
    plt.scatter(el[:, 0], el[:, 1], c=colours[i])
    plt.gca().set_aspect('equal')
    plt.show()
    
## Together Second
for i,el in enumerate(graph["vertex_coord_list"]):
    plt.scatter(el[:, 0], el[:, 1], c=colours[i])
    plt.gca().set_aspect('equal')
plt.show()
```
