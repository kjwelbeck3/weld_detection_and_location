```python
def plot_pc_with_label_overlay(pc, label, destination_file=None):

    fig, ax = plt.subplots(1, 1, figsize=(1.5, 3.5), facecolor='none', tight_layout=True)  
    # pc.draw_normals(fig)
    
    ## drawing each point
    points = ax.scatter(pc[:,0], pc[:,1], s=5.0, c="#9898a3")
    
    ## drawing staple overlay at label location
    markers = add_staple_patch(ax, label[0], label[1], 0, label[2])
    
    ## drawing center point
    center = ax.scatter(label[0], label[1], marker="X", c="#000000")
    
    ax.set_title(f"{label[0], label[1]}")
    
    ## saving to file
    if destination_file:
        plt.savefig(destination_file)
        plt.close()
        plt.cla()
        plt.clf()
    
def parse_weld_labelfile(file_path):
    with open(file_path, 'r') as labelfile:
        line = labelfile.readline()
        x, y, cls = line.split()
    return (float(x), float(y), int(cls))
    
```

```python
## Looping through and plotting the per weld point clouds with label overlay

pc_files = os.listdir(point_clouds_dir)
for i, file in enumerate(pc_files):
    tag = file.strip("npy")
    pc = np.load(point_clouds_dir+file)
    label = parse_weld_labelfile(labels_dir+tag+"txt")

#     print(tag)
#     print(label)
#     print()
#     plot_pc_with_label_overlay(pc, label)

    plot_pc_with_label_overlay(pc, label, imgs_dir+tag+".png")
    print(f"{i+1}/{len(pc_files)}  - DONE")
```
