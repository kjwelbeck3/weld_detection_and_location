```python
######## Settings

### Checking the test set
# weld_pc_dir = "./Recreating Dataset/locationing_dataset/test/point_clouds/"
# labels_dir = "./Recreating Dataset/locationing_dataset/test/labels/"

### Checking the train set
weld_pc_dir = "./Recreating Dataset/locationing_dataset/train/point_clouds/"
labels_dir = "./Recreating Dataset/locationing_dataset/train/labels/"

assert os.path.isdir(weld_pc_dir), "PC directory location does not exist"
assert os.path.isdir(labels_dir), "Labels directory location does not exist"

## Model Parameters
root = "./_loc_model/original_test/"
# model_params_path = root+'2024_01_06_16_00_26/min_test_loss_params_epoch_2.pt'
model_params_path = root+'2024_01_06_16_00_26/params_epoch_0_2024_01_06_16_01_53.pt'
# model_params_path = root+'2024_01_06_16_00_26/params_epoch_1_2024_01_06_16_03_19.pt'
# model_params_path = root+'2024_01_06_16_00_26/min_test_loss_params_epoch_2.pt'
# model_params_path = root+'2024_01_06_16_00_26/min_test_loss_params_epoch_2.pt'

assert os.path.isfile(model_params_path)

## Model Defined:
model = LocationRegressionModelV1()
device = "cpu"


######## Setup
## The Model
model.load_state_dict(torch.load(model_params_path))
model.to(device)
model.eval()

## The Data
dataset = LocDataset(weld_pc_dir, labels_dir)
dataloader = DataLoader(dataset, batch_size=1)
print(f"To process {len(dataloader)} point clouds from {weld_pc_dir}")

with torch.no_grad():
    for i, (X, y) in enumerate(dataloader):
        y = y.to(device)
        y = torch.squeeze(y, 0)

        keys = ("vertex_coord_list", "keypoint_indices_list", "edges_list")
        for key in keys:
            X[key] = [torch.squeeze(tensor,0).to(device) for tensor in X[key]]

        pred = model(X)
        print(pred)
```
