```python
import os
```

```python
## On test split
# input_file = ".\\_data\\test.txt"
# output_file = ".\\_data\\test_v2.txt"

## On train split
input_file = ".\\_data\\train.txt"
output_file = ".\\_data\\train_v2.txt"

if os.path.isfile(input_file):
    with open(input_file, 'r') as ifile:
        with open(output_file, 'w') as ofile:
            lines = ifile.readlines()
            for line in lines:
                scan, label = eval(line)
                scan = scan.replace('./', './_data/')
                label = label.replace('./', './_data/')
                print(scan, label)
                ofile.write(f"['{scan}', '{label}']\n")
            
else:
    print("No such input_file: ", input_file)
            
```
